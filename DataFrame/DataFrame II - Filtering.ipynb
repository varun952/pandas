## and & filtering
### Filter with more than one conditions


```python
mask_1 = df["Salary"] <= 100000
mask_2 = df["Gender"] == "Male"

# filtering below based on two conditions
df[mask_1 & mask_2].head()
```

```python
## or | filtering
### Filter with more than one conditions with or criteria
```

```python
mask_1 = df["Gender"] == "Female"
mask_2 = df["Salary"] <= 100000

df[mask_1 | mask_2].head()
```
