```python
import os
import pandas as pd
os.chdir("/home/mediaworker/anaconda2/envs/pandas_playground/")
df = pd.read_csv("datasets/jamesbond.csv", index_col="Film")
df.sort_index(inplace=True)
df.head(5)
```

```python
# mapper that contains "key" that needs to be
# changes with its mentioned value
mapper = {
    "GoldenEye": "Golden Eye",
    "The World Is Not Enough": "Best Bond Movie Ever"
}

# same results for following calls
df.rename(mapper=mapper, axis=0)
df.rename(mapper=mapper, axis="rows")
df.rename(mapper=mapper, axis="index")
df.rename(index=mapper)

# make permanent
df.rename(index=mapper, inplace=True)

df
```

```python
# change names of column labels
df.rename(mapper = {"Year": "Release Date"}, axis="columns")
df.rename(columns = {"Year": "Release Date"}, inplace=True)
df
```
